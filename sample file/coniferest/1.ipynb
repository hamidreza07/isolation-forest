```python
df=pd.read_csv(r"train.csv",index_col='PassengerId')
```

```python
df=df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']]
```

```python
for colname in df.select_dtypes("object"):
    df[colname], _ = df[colname].factorize()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df.to_numpy()
```

```python
import sklearn.ensemble
import coniferest.isoforest
from coniferest.datasets import MalanchevDataset

# 1e6 data points
dataset = df.to_numpy()

# %%time
isoforest = coniferest.isoforest.IsolationForest(n_subsamples=1024)
isoforest.fit(dataset)
scores = isoforest.score_samples(dataset)
# CPU times: user 16.4 s, sys: 26.1 ms, total: 16.4 s
# Wall time: 5.03 s

# %%time
skforest = sklearn.ensemble.IsolationForest(max_samples=1024)
skforest.fit(dataset)
skscores = skforest.score_samples(dataset)
# CPU times: user 32.3 s, sys: 4.48 s, total: 36.8 s
# Wall time: 36.8 s
```

```python
skscores
```

```python
dataset['score']=scores
dataset['skscores']=skscores
```

```python
dataset.sort_values(by=['score','skscores'],ascending=True)
```

```python
dataset
```

```python
# %%time
from coniferest.sklearn.isoforest import IsolationForestEvaluator
dataset = dataset.drop(['score','skscores'],axis=1)
isoforest = sklearn.ensemble.IsolationForest(max_samples='auto',contamination=float(0.2),random_state=10)


isoforest.fit(dataset[['Survived']])
print(isoforest.get_params())

dataset['scores'] = isoforest.decision_function(dataset[['Survived']])

dataset['anomaly_score'] = isoforest.predict(dataset[['Survived']])



# Wall time: 6.32 s
```
